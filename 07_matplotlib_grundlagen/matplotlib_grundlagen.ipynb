# Eine Testzeitreihe erzeugen

Wir erzeugen mal eine Zeitreihe, bei der jeder Einzelschritt zufällig ist (aber von der gleichen Wahrscheinlichkeitsverteilung stammt) und auf den vorherigen Wert drauf addiert wird.

```python
import random

zr = [0]
N = 100
for i in range(0,N):
    zr.append(zr[i] + random.uniform(-1.0,1.0))
```

```python
zr[0:10]
```

# Ein erster Plot

```python
import matplotlib.pyplot as plt
plt.plot(zr)
plt.show()
```

# Linienfarbe ändern

```python
plt.plot(zr, "green")
plt.show()
```

Für die Spezifikation der Farbe des Graphen über Standardfarben stehen auch folgende Buchstaben als Abkürzung für die Farben zur Verfügung:
    
    b : blue
    g : green
    r : red
    c : cyan
    m : magenta
    y : yellow
    k : black
    w : white

Wir können die Linienfarbe auch über ein RGB-Tupel angeben:

```python
plt.plot(zr, color=(0.0,0.0,1.0))
plt.show()
```

# Linienstyle ändern

Wie die Linie für den Graphen gezeichnet werden soll (durchgehende Linie, gestrichelt, Punkte, etc.) kann auch über die Zeichenkette angegeben werden:

```python
plt.plot(zr, "--")
plt.show()
```

Hierbei stehen folgende Standardlinienarten zur Verfügung:

    '--', '-.', '-', ':'

```python
plt.plot(zr, "-.")
plt.show()
```

```python
plt.plot(zr, ":")
plt.show()
```

Man kann auch beides in der Zeichenkette spezifizieren: Linienstyle und die Farbe:

```python
plt.plot(zr, "k-.") # k steht für "blacK"
plt.show()
```

Man kann den Linienstyle und die Farbe auch explizit über die Schalter `linestyle` und `color` spezifizieren:

```python
plt.plot(zr, linestyle=":", color="green")
plt.show()
```

# Datenpunkte hervorheben

Matplotlib hat die einzelnen Datenpunkte mit Linien verbunden. Vielleicht ist das aber nicht unbedingt sinnvoll (je nach Daten) und wir wollen daher zusätzlich die einzelnen Originaldatenpunkte sehen.

Das geht mit Hilfe des `marker` Schalters.

```python
plt.plot(zr, "-.", color=(0.5,0.5,0.5), marker="o")
plt.show()
```

Auch hier haben wir wieder eine große Auswahl an möglichen Styles für die Marker:

    'o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd'

```python
plt.plot(zr, "k-.", marker="+")
plt.show()
```

Wollen wir nur die Datenpunkte sehen, können wir das so erreichen:

```python
plt.plot(zr, "+", color="black")
plt.show()
```

Jetzt wäre es noch schön, wenn die markierten Datenpunkte sich farblich von der Linie des Graphen unterscheiden. Dazu kann man vor die Spezifikation des Markers `"+"` auch noch einen der folgenden Buchstaben für die folgenden Standardfarben setzen:

```python
plt.plot(zr, "+", color="red") # Plotten der Datenpunkte
plt.plot(zr, linestyle="--", color="gray") # Plotten des Graphen
plt.show()
```

# Logarithmische Skala für die y-Achse

Der y-Wertebereiche ist sehr groß, wenn wir z.B. die Exponentialfunktion `exp()` betrachten.

Bei einer linearen Skala sind der Graph der Funktion in dem Wertebereich zwischen 0 und 10 dann so aus:

```python
import matplotlib.pyplot as plt
import numpy as np

x_min = 0
x_max = 10.0

x = np.arange(x_min, x_max, 0.01)
y = np.exp(x)

plt.plot(x,y, linestyle="-", color="black")

plt.xlim(x_min,x_max)
plt.ylim(np.exp(x_min),np.exp(x_max))

plt.grid(True, linestyle='--')
plt.show()
```

Jetzt aktivieren wir mit `yscale()` einmal einen logarithmischen Wertebereich:

```python
import matplotlib.pyplot as plt
import numpy as np

x_min = 0
x_max = 10.0

x = np.arange(x_min, x_max, 0.01)
y = np.exp(x)

plt.plot(x,y, linestyle="-", color="black")

plt.xlim(x_min,x_max)
plt.ylim(np.exp(x_min),np.exp(x_max))

plt.yscale('log')

plt.grid(True, linestyle='--')
plt.show()
```

# Scatter plot

Deutsche Bezeichnung: "Punktediagramme", "Streudiagramme"

Was ist, wenn unsere Daten zwei-dimensional sind, also wir nicht eine Funktion plotten wollen, die für jeden x-Wert maximal einen y-Wert hat, sondern wir eine Menge von Punkten

    { (x1,y1), (x2,y2), ..., (xn,yn) }
    
plotten wollen?

```python
# Z.B. alter einer Person
# Beachte: es gibt mehrere Personen,
# die das gleiche Alter haben
x_werte = [10,10,12,12,14,15]

# Z.B. Größe einer Person
y_werte = [140,145,160,168,170,175]

plt.scatter(x_werte,y_werte, marker="*", color="blue")

plt.show()
```

# Mehrere Diagramme in einer Figure (Subplots)

```python
fig = plt.figure()

# Wir wollen ein Gitter von 2x2 Plots

# Beachte: Die 3. Zahl gibt die Plot-Nr an
plt1 = fig.add_subplot(2,2,1)
plt1.plot(zr)
plt1.set_title("Zeitreihe als Graph")

plt2 = fig.add_subplot(2,2,2)
plt2.hist(zr)
plt2.set_title("Häufigkeiten der Werte")

plt3 = fig.add_subplot(2,2,3)
plt3.plot(zr, ".")
plt3.set_title("Nur die Datenpunkte")

plt.show()
```

# Abstände zwischen Subplots definieren

Wir brauchen mehr Abstand zwischen den Subplots, damit der Titel des unteren Subplots überhaupt lesbar wird:

```python
plt.clf()

fig = plt.figure()

# Wir wollen ein Gitter von 2x2 Plots

plt1 = fig.add_subplot(2,2,1)
plt1.plot(zr)
plt1.set_title("Zeitreihe als Graph")

plt2 = fig.add_subplot(2,2,2)
plt2.hist(zr)
plt2.set_title("Häufigkeiten der Werte")

plt3 = fig.add_subplot(2,2,3)
plt3.plot(zr, ".")
plt3.set_title("Nur die Datenpunkte")

plt.subplots_adjust(wspace=0.4, hspace=0.6)

plt.show()
```

Was bedeuten die Parameter von `subplots_adjust()`?

    left = 0.125  # the left side of the subplots of the figure
    right = 0.9   # the right side of the subplots of the figure
    bottom = 0.1  # the bottom of the subplots of the figure
    top = 0.9     # the top of the subplots of the figure
    wspace = 0.2  # the amount of width reserved for space between
                  # subplots, expressed as a fraction of the average
                  # axis width
    hspace = 0.2  # the amount of height reserved for space between
                  # subplots, expressed as a fraction of the average
                  # axis height

# 3D Plots

Matplotlib wurde eigentlich entwickelt, um 2D Plots zu erzeugen.

Mit Hilfe des Pakets `mplot3d` ist es aber möglich auch 3D Plots zu erzeugen.

Auf der `mplot3d` Webseite unter

https://matplotlib.org/mpl_toolkits/mplot3d/index.html

heißt es:

>The mplot3d toolkit adds simple 3D plotting capabilities to matplotlib by supplying an axes object that can create a 2D projection of a 3D scene. The resulting graph will have the same look and feel as regular 2D plots.

Das Schöne ist, dass die 3D Plots (außerhalb von Jupyter) auch gleich interaktiv sind:

> The interactive backends also provide the ability to rotate and zoom the 3D scene. One can rotate the 3D scene by simply clicking-and-dragging the scene. Zooming is done by right-clicking the scene and dragging the mouse up and down. Note that one does not use the zoom button like one would use for regular 2D plots

Wir erzeugen nun einen ersten noch leeren 3D Plot:

```python
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection="3d")

plt.show()
```

Wir erzeugen jetzt erstmal eine 3D Zeitreihe:

```python
z_werte = np.linspace(0, 15, 100)
x_werte = np.cos(z_werte)
y_werte = np.sin(z_werte)
```

Jetzt plotten wir die 3D Zeitreihe als Graphen:

```python
plt.clf()

fig = plt.figure()
ax = plt.axes(projection="3d")
print(type(ax))

ax.plot3D(x_werte, y_werte, z_werte, 'gray')

ax.set

plt.show()
```

Einige Achsenwerte erscheinen etwas stark gedrängt. Wir können ja wieder die xticks und yticks selber setzen:

```python
plt.clf()

ax = plt.axes(projection="3d")
print(type(ax))

ax.plot3D(x_werte, y_werte, z_werte, 'gray')

ax.set_xticks( np.arange(-1.0, 1.01, 0.5) )
ax.set_yticks( np.arange(-1.0, 1.01, 0.5) )

plt.show()
```

Achsentitel:

```python
plt.clf()

ax = plt.axes(projection="3d")
print(type(ax))

ax.plot3D(x_werte, y_werte, z_werte, 'gray')

ax.set_xticks( np.arange(-1.0, 1.01, 0.5) )
ax.set_yticks( np.arange(-1.0, 1.01, 0.5) )

ax.set_xlabel("x", fontsize=15, color ="red")
ax.set_ylabel("y", fontsize=15, color ="green")
ax.set_zlabel("z", fontsize=15, color ="blue")

plt.show()
```

Datenpunkte miteinzeichnen (Scatter plot in 3D):

```python
plt.clf()

ax = plt.axes(projection="3d")
print(type(ax))

ax.plot3D(x_werte, y_werte, z_werte, 'gray')

ax.set_xticks( np.arange(-1.0, 1.01, 0.5) )
ax.set_yticks( np.arange(-1.0, 1.01, 0.5) )

ax.set_xlabel("x", fontsize=15, color ="red")
ax.set_ylabel("y", fontsize=15, color ="green")
ax.set_zlabel("z", fontsize=15, color ="blue")

ax.scatter3D(x_werte, y_werte, z_werte, color="green");

plt.show()
```

Ansichtswinkel definieren:

```python
plt.clf()

ax = plt.axes(projection="3d")
print(type(ax))

ax.plot3D(x_werte, y_werte, z_werte, 'gray')

ax.set_xticks( np.arange(-1.0, 1.01, 0.5) )
ax.set_yticks( np.arange(-1.0, 1.01, 0.5) )

ax.set_xlabel("x", fontsize=15, color ="red")
ax.set_ylabel("y", fontsize=15, color ="green")
ax.set_zlabel("z", fontsize=15, color ="blue")

ax.scatter3D(x_werte, y_werte, z_werte, color="black");

ax.view_init(30, 90)

plt.show()
```

Aus der Doku zu `mplt3d` wird die Bedeutung der beiden Parameter von `view_init()` ersichtlich:

>Set the elevation and azimuth of the axes in degrees (not radians).

>This can be used to rotate the axes programmatically.

>'elev' stores the elevation angle in the z plane (in degrees). 'azim' stores the azimuth angle in the x,y plane (in degrees).

Damit können wir eine kleine Animation machen:

```python
plt.clf()

fig = plt.gcf()

ax = plt.axes(projection="3d")
print(type(ax))

ax.plot3D(x_werte, y_werte, z_werte, 'gray')

ax.set_xticks( np.arange(-1.0, 1.01, 0.5) )
ax.set_yticks( np.arange(-1.0, 1.01, 0.5) )

ax.set_xlabel("x", fontsize=15, color ="red")
ax.set_ylabel("y", fontsize=15, color ="green")
ax.set_zlabel("z", fontsize=15, color ="blue")

ax.scatter3D(x_werte, y_werte, z_werte, color="black");

for angle in range(0,90):
    ax.view_init(angle/2, angle)
    fig.savefig("animation/view{0:03d}.png".format(angle))
    plt.show()
```

Mit Tools wie z.B. ffmpeg kann man dann aus den weggeschriebenen Einzelbildern leicht ein animated gif oder ein Video erstellen:

    ffmpeg -i view%03d.png -qscale 0 animation.mp4
    
oder

    ffmpeg -i view%03d.png -qscale 0 animation.gif
